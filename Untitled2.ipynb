```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
data["Age"].fillna(data["Age"].mean(), inplace=True)
data["Height"].fillna(data["Height"].mean(), inplace=True)
data["Weight"].fillna(data["Weight"].mean(), inplace=True)
data["BMI"].fillna(data["BMI"].mean(), inplace=True)
data["Temperature"].fillna(data["Temperature"].median(), inplace=True)
data["Heart Rate"].fillna(data["Heart Rate"].mean(), inplace=True)
data["Blood Pressure"].fillna(data["Blood Pressure"].mean(), inplace=True)
data["Cholesterol"].fillna(data["Cholesterol"].median(), inplace=True)
```

```python
for i in ['Gender','Blood Type','Diabetes']:
    data[i].fillna(method="ffill", inplace=True)

data['Smoking'].fillna(method="bfill", inplace=True)


data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']
le = LabelEncoder()

for col in label_cols:
    data[col] = le.fit_transform(data[col])

data
```

```python
numeric_cols = ['Age', 'Height', 'Weight', 'BMI', 'Temperature',
                'Heart Rate', 'Blood Pressure', 'Cholesterol']


def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Check how many outliers are in each column
for col in numeric_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

```python
numerical_cols = [ 'BMI', 'Blood Pressure']

plt.figure(figsize=(12, 6))
data[numerical_cols].boxplot()
plt.title("Boxplot Before Outlier Removal")
plt.xticks(rotation=45)
plt.show()
```

```python

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

for col in numeric_cols:
    data = remove_outliers_iqr(data, col)
```
